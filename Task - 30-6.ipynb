 # Merging or Splitting PDFs

```python
from PyPDF2 import PdfMerger

merger = PdfMerger()
merger.append(r"C:\Users\Chetan Habib\Downloads\PROJECT SYNOPSYS.pdf")
merger.append(r"C:\Users\Chetan Habib\Downloads\AI- INVESTMENT ADVISOR REPORT.pdf")
merger.write("merged.pdf")
merger.close()
merger
```

```python
import PyPDF2

with open("merged.pdf", "rb") as file:
    reader = PyPDF2.PdfReader(file)
    for page in reader.pages:
        text = page.extract_text()
        print(text)
```

```python
from PyPDF2 import PdfMerger

merger = PdfMerger()
merger.append(r"C:\Users\Chetan Habib\Downloads\MySQL Interview Questions with Answers by Jeevan Raj (1).pdf")
merger.append(r"C:\Users\Chetan Habib\Downloads\Statistic Interview Questions and Answers by Jeevan Raj.pdf")
merger.write("merged1.pdf")
merger.close()
merger
```

```python
import PyPDF2

with open("merged1.pdf", "rb") as file:
    reader = PyPDF2.PdfReader(file)
    for page in reader.pages:
        text = page.extract_text()
        print(text)
```
